```python
import cv2
import matplotlib.pyplot as plt
```

```python
#read image from storage
img = cv2.imread('/home/fayis/Desktop/me/image_processing/images/Lenna.png')
#convert to RGB channel
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img,)
```

```python
#function to convert to negative image
def negative(img):
    return 255-img
#function to convert to negative image using open cv
def negative_using_openCV(img):
    return cv2.bitwise_not(img)
```

```python
negative_img= negative(img)
plt.imshow(negative_img)
```

```python
cv_negative_image = negative_using_openCV(img)
plt.imshow(cv_negative_image)
```
