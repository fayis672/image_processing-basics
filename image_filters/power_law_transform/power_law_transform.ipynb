```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
#read image from storage
img = cv2.imread('/home/fayis/Desktop/me/image_processing/images/Lenna.png')
#convert to RGB channel
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img,)
```

```python
# power law transform with two gamma levels
def power_law_transform(img,gamma):
    return np.array(255*(img/255)**gamma,dtype=np.uint8)

gamma1= 0.25
gamma2 = 3.0
gamma1_img = power_law_transform(img,gamma1)
gamma2_img = power_law_transform(img,gamma2)
cmp =np.hstack((img,gamma1_img,gamma2_img))
plt.imshow(cmp)    
```
